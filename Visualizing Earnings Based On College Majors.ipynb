# Visualizing Earnings Based On College Majors

We'll be visualizing a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.


```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import dataframe in pandas
recent_grads = pd.read_csv('recent-grads.csv')
```

#### Explore dataframe

```python
#Original dataframe length
raw_data_count =len(recent_grads)
```

```python
#Drop NA rows
recent_grads = recent_grads.dropna()
```

```python
#Look up number of rows again
print({'initial length':raw_data_count, 'length_without_NA':len(recent_grads)})
```

#### Scatter plots

```python
#Scatter plots
ax = recent_grads.plot(x= 'Sample_size', y = 'Median', kind='scatter')
ax.set_title("Sample size vs. Median")
```

```python
ax = recent_grads.plot(x ='Sample_size',y ='Unemployment_rate',kind = 'scatter')
ax.set_title("Sample_size vs. Unemployment rate")
```

```python
ax = recent_grads.plot('Full_time', 'Median',kind = 'scatter')
ax.set_title('Fulltime vs. Median')
```

```python
ax = recent_grads.plot('ShareWomen','Unemployment_rate', kind = 'scatter')
ax.set_title('Sharewomen vs. Unemployment_rate')
```

```python
recent_grads.plot('Men','Median', kind = "scatter", title ='Men vs. Median')
```

```python
recent_grads.plot(x='Women',y='Median',kind='scatter',title='Women vs. Median')
```

 - Students in more popular majors do not earn more money.
 - Students that major in subjects dominated by women do not earn more money.
 - No link exist between number of full-time employees and median salary.
